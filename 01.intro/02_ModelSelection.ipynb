## Model Selection 모듈

- 훈련데이터와 테스트데이터로 분리하지 않고 머신러닝 수행시

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
dtc = DecisionTreeClassifier()
dtc.fit(iris.data, iris.target)
dtc.score(iris.data, iris.target)
```

- 교차 검증 : Cross Validate() + 평가
- 교차 검증후 평가까지 : cross_val_score()

```python
from sklearn.model_selection import cross_val_score
              # 모델,    x값,       y값                           5번 검증 하겠다 
cross_val_score(dtc, iris.data, iris.target, scoring = "accuracy", cv =5)
```

```python
cross_val_score(dtc, iris.data, iris.target, scoring = "accuracy", cv =5).mean()
```

- GridSearchCV 클래스 : 하이퍼 파라미터 튜닝 + 교차검증

```python
dtc.get_params()
```

```python
# Train/Test dataset 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, stratify = iris.target, random_state = 2022, test_size = 0.2
)
```

```python
# 분류기와 그에 해당하는 하이퍼 파라미터
dtc = DecisionTreeClassifier(random_state = 2022)
dtc.get_params()
```

```python
params = {
    "max_depth":[2,3,4,5,6],
    "min_samples_split" : [2,3,4]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_dt = GridSearchCV(      # 숫자가 큰게 좋은거, 5번 교차 검증 하자
    dtc, param_grid = params, scoring = "accuracy", cv = 5 
)
```

```python
grid_dt.fit(X_train,y_train)
```

```python
grid_dt.score(X_test, y_test)
```

```python
# 베스트 파라메터
grid_dt.best_params_
```

```python
# 베스트 스코어
grid_dt.best_score_
```

```python
# 최적의 분류기
# dtc_best = DecisionTreeClassifier(max_depth = 3, min_samples_split = 2, random_state = 2022)
dtc_best = grid_dt.best_estimator_
```

```python
# 최적의 분류기로 예측 및 평가
dtc_best.score(X_test,y_test)
```

- SVM사례


```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc = SVC(random_state = 2022)
svc.get_params()
```

```python
params = {"C":[0.01,0.1,1,10,100]}
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
```

```python
grid_sv.fit(X_train, y_train)
```

```python
grid_sv.best_params_
```

```python
# 범위를 좁혀가면서 파라메터를 찾음
params = {"C":[0.2,0.6,1,4,8]}
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
grid_sv.fit(X_train, y_train)
grid_sv.best_params_
```

```python
# 범위를 좁혀가면서 파라메터를 찾음
params = {"C":[2,3,4,5,6]}
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
grid_sv.fit(X_train, y_train)
grid_sv.best_params_
```

```python
 svc_best = grid_sv.best_estimator_
 svc_best.score(X_test, y_test)
```
