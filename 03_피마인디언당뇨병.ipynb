```python
df = pd.read_csv("/content/drive/MyDrive/multi/pima-indians-diabetes.csv", skiprows = 9, header = None)
df
```

```python
# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
df.columns = ['P',"G","BP","S","I","BMI","D","Age","Class"]
df.head()
```

```python
# 결측치 확인
df.isna().sum()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X.shape, y.shape
```

- Train/ Test dataset으로 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify = y, test_size = 0.2, random_state = 2022
)
```

```python
np.unique(X_train, return_counts = True)
```

```python
np.unique(y_test, return_counts = True)
```

- DecisionTreeClassifier


```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 2022)
dtc.get_params()
```

```python
params = {"max_depth":[2,3,4,5,6],
          "min_samples_split":[2,3,4]}
```

```python
from sklearn.model_selection import GridSearchCV

grid_dt = GridSearchCV(dtc, params, scoring="accuracy", cv = 5)
grid_dt.fit(X_train, y_train)
```

```python
grid_dt.best_params_
```

```python
best_dt = grid_dt.best_estimator_
best_dt.score(X_test, y_test)
```
