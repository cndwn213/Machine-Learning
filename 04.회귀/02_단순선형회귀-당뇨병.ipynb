## 단순 선형 회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- 당뇨병 데이터 셋 탐색

```python
from sklearn.datasets import load_diabetes
dia = load_diabetes()
```

```python
df = pd.DataFrame(dia.data, columns = dia.feature_names)
df["target"] = dia.target
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
_, axs = plt.subplots(2, 5, figsize = (18, 8))
for i, feature in enumerate(df.columns[:-1]):
  row, col = i // 5, i % 5
  sns.regplot(x = feature, y = "target", data = df, ax = axs[row][col])
```

- feature vs target

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
for feature in df.columns[:-1]:
  X = df[feature].values.reshape(-1,1)
  lr.fit(X, df.target)
  score = lr.score(X, df.target)
  print(f"{feature} : {score : .4f}")
```

### BMI vs target
- 훈련/테스트 데이터셋 분리(test_size = 0.1)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df["bmi"].values.reshape(-1,1), df["target"], test_size = 0.1,random_state = 1018
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
# 계수,      절편,       잔차제곱합
lr.coef_, lr.intercept_, lr._residues
```

```python
# R squared
from sklearn.metrics import r2_score
pred = lr.predict(X_test)
r2_score(y_test, pred)
```

```python
# Mean Squard Error       # feature 간에 비교할때 사용, 분포도를 비교할 순 없다
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```

- 시각화

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel("bmi")
plt.ylabel("target")
plt.title("Bmi vs target", fontsize = 20)
plt.show()
```

```python
xs = np.array([-0.1, 0.18])
ys = xs * lr.coef_ + lr.intercept_
ys
```

```python
plt.figure(figsize = (10,8)) # figure가 제일 처음에 나와야 한다.
plt.scatter(X_train, y_train, label = "train set")
plt.scatter(X_test, y_test, marker = "^" ,label = "test set")
plt.plot(xs, ys, "r-", lw = 3)
plt.legend()
plt.grid()
plt.xlabel("bmi")
plt.ylabel("target")
plt.title("Bmi vs target", fontsize = 20)
plt.show()
```
