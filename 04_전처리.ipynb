# 데이터 전처리

### 1. 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder
```

```python
items = ["TV", "냉장고", "전자렌지", "컴퓨터", "선풍기", "선풍기", "믹서", "믹서"]
```

```python
# LabelEncoder 객체 생성
le = LabelEncoder()
```

```python
# 학습할 데이터를 훑어 봄
le.fit(items)
```

```python
# 인코딩 실행
labels = le.transform(items)
labels
```

```python
# 실전에서는
le2 = LabelEncoder()
labels = le2.fit_transform(items)
labels
```

```python
# 고수들은?
labels = LabelEncoder().fit_transform(items)
labels
```

```python
le.inverse_transform([0,1,2,3,4,5])   # decoding
```

### 2. One-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
oh_labels = ohe.fit_transform(labels.reshape(-1,1))
oh_labels.toarray()
```

```python
from tensorflow.keras.utils import to_categorical

to_categorical(labels)
```

### 3. 표준화
- 평균 = 0, 표준편차 1인 가우시안 표준정규분표

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
from sklearn.preprocessing import StandardScaler

iris_std = StandardScaler().fit_transform(iris.data)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.describe()
```

```python
df = pd.DataFrame(iris_std, columns = iris.feature_names)
df.describe()
```

- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state = 2022)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain,ytest = train_test_split(
    iris.data, iris.target, stratify = iris.target, random_state = 2022
)
lrc.fit(Xtrain,ytrain)
```

```python
# 표준정규분포(sklearn.preprocessing.StandardScaler)로 변환된 데이터로 학습하면 
# 빨리 목표값에 도달해서 에러가 발생하지 않음
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain,ytest = train_test_split(
    iris_std, iris.target, stratify = iris.target, random_state = 2022
)

lrc.fit(Xtrain,ytrain)
```

```python
lrc.score(Xtest,ytest)
```

### 4. 정규화
- 최소값 0, 최대값 1 로 변환

```python
from sklearn.preprocessing import MinMaxScaler

iris_mm = MinMaxScaler().fit_transform(iris.data)
```

```python
df3 = pd.DataFrame(iris_mm, columns = iris.feature_names)
df3.describe()
```

```python
from sklearn.linear_model import LogisticRegression
# 정규화(sklearn.preprocessing.MinMaxScaler) 변환된 데이터로 학습하면 
# 빨리 목표값에 도달해서 에러가 발생하지 않음
Xtrain, Xtest, ytrain,ytest = train_test_split(
    iris_mm, iris.target, stratify = iris.target, random_state = 2022
)
lrc = LogisticRegression(random_state = 2022)
lrc.fit(Xtrain, ytrain)
```

```python
lrc.score(Xtest,ytest)
```
