### 1. Adjacent Matrix 생성
- 행은 사용자
- 열은 영화
- 내용은 평점

```python
raw_data = np.array(data.raw_ratings, dtype = int)
np.min(raw_data, axis=0),np.max(raw_data, axis=0) 
```

```python
df = df.astype(int)
df["user-id"].min(),df["user-id"].max()
```

```python
# user-id, movie-id가 0부터 시작하도록 만들어 줌
raw_data[:, :2] -= 1
raw_data[:5]
```

```python
# 행, 열의 갯수
n_rows = df["user-id"].nunique() # 행의 갯수 = 사용자의 수
n_cols = df["movie-id"].nunique() # 열의 갯수 = 영화 갯수
n_rows,n_cols
```

#### 1) 본 영화/ 안본 영화로만 구분, 0/1

```python
adj_matrix = np.zeros([n_rows,n_cols], int)
for user_id, movie_id, _, _ in raw_data:
    adj_matrix[user_id,movie_id] = 1
adj_matrix[:5]    
```

```python
# 0번 사용자가 나라고 함
my_id, my_vector = 0, adj_matrix[0]
```

```python
# 유사도 - 이진 벡터의 내적
# 나와 10번,20번 사용자와의 유사도
np.dot(my_vector, adj_matrix[10]),np.dot(my_vector, adj_matrix[20])
```

```python
# 누가 나랑 닮았나?
best_score, best_match_id = 0,0
for i in range(len(adj_matrix)):
    if i == 0:
        continue
    dot = np.dot(my_vector, adj_matrix[i])
    if best_score < dot:
        best_score ,best_match_id = dot, i
best_score ,best_match_id 
```

```python
best_vector = adj_matrix[best_match_id]
my_vector[200 :210], best_vector[200 :210]
```

```python
#  내가 보지 않은 영화중에서 best+match 사용자가 본 영화 --> 추천
recommend_list = []
for i, (my_view, best_match_view) in enumerate(zip(my_vector, best_vector)):
    if my_view == 0 and best_match_view == 1 : 
        recommend_list.append(i)

len(recommend_list)
```

### 2) 평점 점수를 주는 경우

```python
adj_matrix = np.zeros([n_rows,n_cols], int)
for user_id, movie_id, rating, _ in raw_data:
    adj_matrix[user_id,movie_id] = rating
adj_matrix[:5]    
```

- Case 1: 유클리드 거리

```python
# 누가 나랑 닮았나?
best_score, best_match_id = 10000,0
my_vector = adj_matrix[0]
for i in range(1, len(adj_matrix)):
    euc = np.sqrt(np.sum(np.square(my_vector - adj_matrix[i]))) # 유클리드 거리 공식
    if best_score > euc:
        best_score ,best_match_id = euc, i
best_score ,best_match_id 
```

```python
#  내가 보지 않은 영화중에서 best_match 사용자가 평점 5점 준 영화 --> 추천
best_vector = adj_matrix[best_match_id]
recommend_list = []
for i, (my_view, best_match_view) in enumerate(zip(my_vector, best_vector)):
    if my_view == 0 and best_match_view == 5 : 
        recommend_list.append(i)

len(recommend_list), recommend_list[:10]
```

```python
#  내가 보지 않은 영화중에서 best_match 사용자가 본 평점이 좋은 영화 --> 추천
best_vector = adj_matrix[best_match_id]
recommend_list = []
for i, (my_view, best_match_view) in enumerate(zip(my_vector, best_vector)):
    if my_view == 0 and best_match_view >= 4 : 
        recommend_list.append(i)

len(recommend_list), recommend_list[:10]
```

- Case 2: 코사인 유사도

```python
def cosine_similarity(v1, v2):
    v1_norm = np.sqrt(np.sum(np.square(v1)))
    v2_norm = np.sqrt(np.sum(np.square(v2)))
    return np.dot(v1, v2) / (v1_norm * v2_norm)
```

```python
# 누가 나랑 닮았나?
best_score, best_match_id = -1, 0
my_vector = adj_matrix[0]
for i in range(1, len(adj_matrix)):
    cos = cosine_similarity(my_vector, adj_matrix[i]) # 코사인 유사도
    if best_score < cos:
        best_score ,best_match_id = cos, i
best_score ,best_match_id 
```

```python
#  내가 보지 않은 영화중에서 best_match 사용자가 본 평점이 좋은 영화 --> 추천
best_vector = adj_matrix[best_match_id]
recommend_list = []
for i, (my_view, best_match_view) in enumerate(zip(my_vector, best_vector)):
    if my_view == 0 and best_match_view >= 4 : 
        recommend_list.append(i)

len(recommend_list), recommend_list[:10]
```
