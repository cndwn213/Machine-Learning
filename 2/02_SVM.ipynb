# SVM(Support Vector Machine)

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
import pandas as pd
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
df["target"] = cancer.target
print(df.shape)
df.head(3)
```

```python
# y  
df.target.value_counts()
```

- 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cancer_std = scaler.fit_transform(cancer.data)
df = pd.DataFrame(cancer_std, columns = cancer.feature_names)
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer_std, cancer.target, stratify = cancer.target, random_state = 2022, test_size = 0.2
)
```

```python
X_train.shape, y_train.shape
```

- SVM  ,   

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
svc.score(X_test, y_test)
```

```python
svc.get_params()
```

### Toy MNIST data

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
digits.keys()
```

```python
digits.images.shape
```

```python
digits.data.min(), digits.data.max()
```

```python
import matplotlib.pyplot as plt
plt.imshow(digits.images[0])
plt.show()
```

```python
plt.figure(figsize = (1,1))
plt.imshow(digits.images[0], cmap = plt.cm.binary, interpolation = "nearest")
plt.axis(False)
plt.show()
```

```python
plt.figure(figsize = (10,1))
for i in range(10):
  plt.subplot(1, 10, i + 1)
  plt.imshow(digits.images[i], cmap = plt.cm.binary, interpolation = "nearest")
  plt.axis(False)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    digits.data/16, digits.target, stratify = digits.target,
    random_state = 2022, test_size = 0.2
)
```

- svm ,   

```python
params = {"C":[0.01,0.1,1,10,100]}
```

```python
from sklearn.model_selection import GridSearchCV
svc = SVC(random_state = 2022)
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
grid_sv.fit(X_train, y_train)
grid_sv.best_params_
```

```python
params = {"C":[2,6,10,50]}
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
grid_sv.fit(X_train, y_train)
grid_sv.best_params_
```

```python
params = {"C":[9]}
grid_sv = GridSearchCV(svc, params, scoring = "accuracy", cv = 5)
grid_sv.fit(X_train, y_train)
grid_sv.best_params_
```

```python
best_sv = grid_sv.best_estimator_
best_sv.score(X_test, y_test)
```

```python
svc = SVC(random_state = 2022)
svc.fit(X_train, y_train)
svc.score(X_test, y_test)
```
